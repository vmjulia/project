```python
# in addition to cluster and classification data, search for key wordsby defining a regex pattern
def find_groups(text):
    group_words = [
        'gruppe',
        'grupo',  
        'groupe', 
        'gruppo', 
        'group',
        'team',
        "zu zweit",
        'équipe',  
        'equipe',  
        'squadra',  
        'equipo',  
    ]

    # Combine group words into a single regex pattern
    group_pattern = r'(?:' + '|'.join(re.escape(word) for word in group_words) + r')'

    matches = re.findall(group_pattern, text, flags=re.IGNORECASE)  # Using re.IGNORECASE to make the search case-insensitive

    return matches

# this is the main function which uses clustering, classification and key word search results and allows for improve,emt by providing user input
# if the key words are matched or classification model identified preparation in teams with 70 percent prob or the group work cluster is idenrtify 
# user is supposed to enter 1 to have True for group work, 0 to have False. This code can be modified and used without user input to save time e.g. by just always setting true if 1 of the conditions is satisfied 
# however for higher accuracy to exclude false positives I use user input 8 it is not too long, about an hour for the whole data that I processed
# the results are stored in step_2 csv file
def process_data(year, period):
    ind = str(year) + ("00") + str(period)
    df = pd.read_csv('data/' + 'step_1_%s.csv' % ind) 
    for idx, row in df.iterrows():
            words = []
                            
            # Check exam data
            if not pd.isna(row["Assessment details"]):
                words = find_groups(row["Assessment details"])
                
            # Get unique words list
            words = list(set(words))
            
            if len(words) > 0 or df.loc[idx,"preparation in teams"]>0.7  or df.loc[idx,"goup_work_cluster"]:
                print("score:", df.loc[idx,"preparation in teams"])
                print("clustering:", df.loc[idx,"goup_work_cluster"])
                print("words:", words)
                print("Assessment details:", row["Assessment details"])
                
                user_input = input('Enter 1 to have True for group work, 0 to have False ')
                if user_input.isdigit():
                    number = int(user_input)
                    if number == 0:
                        df.loc[idx,"goup_work"] = False
                    elif number == 1:
                        df.loc[idx,"goup_work"] = True
                    else:
                        print("Invalid input.")
                else:
                    print("Invalid input.")

    df.to_csv('data/' + 'step_2_%s.csv' % ind,  encoding='utf-8', index=False)  
    

# Iterate over years and periods
for year in range(2016, 2023):
    for period in range(4, 6):
        process_data(year, period)
```

# STEP 3:  process programming languages

```python
import re
import pandas as pd
# R and C have some false positives... so I excluded some cases with this pattern
# therefor they have a separate function "find_programming_languages" in addition to "find_programming_languages_2"
# find_programming_languages is stricter and excludes many false posirives

def find_programming_languages(text):
    programming_languages = [
        'R',
        "C"
    ]
    
    # Combine  programming languages into a single regex pattern
    pattern = r'(?<![-.])\b(?!(?:C\+\+))\b(?:' + '|'.join(programming_languages) + r')\b(?!-)(?![C.])\b(?!&\s*D)'

    matches = re.findall(pattern, text)

    return matches

# other programming languages can be searched case-insesitively and the condition for regex is less strict 
def find_programming_languages_2(text):
    programming_languages = [
        'Python',
        'Java',
        'R-studio',
        'C#',
        'JavaScript',
        'Ruby',
        'Swift'
        'PHP',
        'TypeScript',
        'Kotlin',
        'Perl',
        'Scala',
        'Rust',
        'Haskell',
        'SQL',
        'HTML',
        "HTML5",
        "XML",
        'Matlab',
        'Bash',
        "RStudio",
        "Linux",
        "SPSS",
        "Mplus",
        "Simulink",
        "Unix",
        "Golang",
        "Fortran",
        "Groovy",
        "Latex",
        "Stata"
        
    ]
    # Combine all programming languages into a single regex pattern
    pattern = r'(?<!-)(?<!\.)\b(?:' + '|'.join(programming_languages) + r')\b'

    matches = re.findall(pattern, text, flags = re.IGNORECASE)
    return matches

# Example usage:
text = "I love Python and JavaScript, but I'm learning C++. Java is also interesting."
languages_mentioned = find_programming_languages_2(text)
print(languages_mentioned)
```

```python
# This is the main loop for programming languages, I try to find the key words. R and C have some fals positives still. Thetrefore I still get a user input for these ones
# the lanuages missing: Assembly, Go. I verified all the cases and these 2 words were never used as programmin lanaguages 
# assembly was always smth in chemistry and Go as go. Therefore, those languages are not matched
# the results of this procedure are stored in step_3 csv.

def process_data(year, period):
    number = str(year) + ("00") + str(period)

    df = pd.read_csv('data/' + 'step_2_%s.csv' % number)
    df['programming_languages'] = [[] for _ in range(len(df))]
    df['programming_languages_flag'] = False
    
    for idx, row in df.iterrows():
        try:
            languages = []
            
            # Check content
            if not pd.isna(row["Content"]):
                res = find_programming_languages_2(row["Content"]) + find_programming_languages(row["Content"])
                if re.search(r'C\+\+', row["Content"]):
                    res.append('C++')
                languages = languages + res
                
            # Check exam data
            if not pd.isna(row["Assessment details"]):
                res = find_programming_languages_2(row["Assessment details"]) + find_programming_languages(row["Assessment details"])
                if re.search(r'C\+\+', row["Assessment details"]):
                    res.append('C++')
                languages = languages + res
                
            # Get unique languages list
            languages = list(set(languages))
            
            if len(languages) > 0:
                print("Languages:", languages)
                print("Content:", row["Content"])
                print("Assessment details:", row["Assessment details"])
                
                user_input = input('Enter "y" to keep all, number to remove an element: ')
                if user_input == 'y':
                    print("kept all")
                    df.loc[idx, "programming_languages"].extend(languages)
                    df.loc[idx, "programming_languages_flag"] = True
                elif user_input.isdigit():
                    index_to_remove = int(user_input)
                    if 0 <= index_to_remove < len(languages):
                        languages.pop(index_to_remove)
                        print("whats left:", languages)
                        df.loc[idx, "programming_languages"].extend(languages)
                        df.loc[idx, "programming_languages_flag"] = len(languages)>0
                    else:
                        print("Invalid input. Please enter a valid number.")
                else:
                    print("Invalid input. Please enter 'y' to keep all or a number to remove an element.")
                    
        except Exception as ex:
            if not pd.isna(row["Assessment details"]):
                print("Exception occurred:", ex, row["Assessment details"])
                break
    df.to_csv('data/' + 'step_3_%s.csv' % number,  encoding='utf-8', index=False)  
    

# Iterate over years and periods
for year in range(2014, 2017):
    for period in range(4, 6):
        process_data(year, period)
```
