```python
# this is notebook, which given the scraped data and the participation data (provided)  
# constructs data of the courses which were attended by students
# the resulting files are named student courses
```

```python
# import libraries
import pandas as pd
import os
```

```python

# Set the directory path where the scraped data is stored
dirr = os.getcwd() + "/test/scraped"  # fixme: Replace with the correct directory path

for year in range(2007, 2023):
    for period in range(4, 6):
        # Read the files with student participation (there were 2 for Umfrage 2021HS and FS 2022)
        df1 = pd.read_csv('participation1.csv')
        df2 = pd.read_csv('participation2.csv')
        
        # Append the two participation dataframes together
        df = df1.append(df2, ignore_index=True)
        
        # Take the required subset of columns
        df_part = df[["Awobjid", "Acad_jahr_buchung", "Acad_periode_buchung"]]
        
        # Select the necessary time period
        d = df_part[df_part["Acad_jahr_buchung"] == year]
        d = d[d["Acad_periode_buchung"] == period]
        
        # Remove duplicates from the selected subset
        final = d.drop_duplicates()
        final["Awobjid"] = final["Awobjid"].astype(str)
        
        ind = str(year) + ("00") + str(period)
        
        # Read the initial scraped data from the teacher training program ( just any to read first and then add the rest in the loop)
        df = pd.read_csv('test/scraped/' + 'courses_%s_Teacher training program_.csv' % ind)
        df.dropna(subset="full_name", inplace=True)
        df["Awobjid"] = df["full_name"].apply(lambda x: x.split("-")[0])
        df_res = df
        
        # Iterate over all the files with scraped data
        for filename in os.listdir(dirr):
            if 'courses_%s' % ind in filename:
                df = pd.read_csv('test/scraped/' + filename)
                df.dropna(subset="full_name", inplace=True)
                df["Awobjid"] = df["full_name"].apply(lambda x: x.split("-")[0])
                df_res = df_res.append(df, ignore_index=True)
        
        # Remove duplicates from the combined scraped data
        df_res.drop_duplicates(subset=['full_name'], inplace=True)
        
        # Save the intermediary step of all scraped data
        df_res.to_csv('test/' + 'all_info_%s.csv' % ind, encoding='utf-8', index=False)
        
        # Check if required columns exist in the scraped data and select only those
        if "Assessment details" in df.columns and "Content" in df.columns:
            df_res = df_res[["full_name", "Content", "Assessment details", "Awobjid"]]
        else:
            print("Columns are different", year)
        
        # Merge the final participation data with the scraped data based on 'Awobjid'
        result = pd.merge(final, df_res, on='Awobjid', how='left')
        
        # Remove duplicates based on the 'Content' and 'Awobjid' columns
        result = result.drop_duplicates(subset=["Content", "Awobjid"])
        
        print(ind, len(result))
        
        # Save the final result of student courses
        result.to_csv('test/' + 'student_courses_%s.csv' % ind, encoding='utf-8', index=False)

    
    
```
